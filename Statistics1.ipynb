
This function defines a Gaussian Distribution (equation 4.3 in the textbook). It makes use of a few new numpy functions: np.sqrt (Square Root), np.exp (Exponential or e^x), and np.pi (simply the constant Pi). 

We then define x using another numpy function, np.arange, which creates an array of values ranging from the minimum to the maximum in increments of 1, which is the default.

If any of these functions are confusing, add in a print command to see what the outputs look like.

We can define y and z as the outputs of the Gaussian function with different inputs.

```python
def Gaussian(xvalues, mean, standard_deviation):
    sdsquared = standard_deviation**2
    exponent = -(xvalues - mean)**2/(2 * sdsquared)
    preexponential = 1 / np.sqrt(2 * np.pi * sdsquared)
    return preexponential * np.exp(exponent)

x = np.arange(-10, 10)
y = Gaussian(x, 0, 3)
z = Gaussian(x, 0, 1)

plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

Gaussian Distributions are defined by two parameters: the mean and the standard deviation. The mean specifies where the center of the distribution is. The standard deviation specifies the width of the peak.

The above block of code also shows you how to plot multiple things on the same graph by just adding extra plot commands before you call show().

This plot shows an example of two Gaussian distributions with the same mean but different standard deviations. Notice, the area under both curves should be 1, which is why the peak intensities are different.

A lot of analytical data take the form of a Gaussian, so it is important that we can calculate these parameters. 

Mean = Sum of X / number of data points in X

Let's calculate the mean using a few programmatic tricks. Here, np.sum() allows us to quickly sum the array and len() lets us calculate how long the array is (the number of data points).

Numpy (of course) has built in functions to calculate mean with np.mean or np.average.

```python
x = np.arange(0, 7) #[0,1,2,3,4,5,6]

mean = np.sum(x)/len(x)
print(mean)

mean2 = np.mean(x)
mean3 = np.average(x)

print(mean2, mean3)
```

These should all give the same output: 3. Make sure you also understand how to do that by hand and compare with equation 4-1 in the textbook.

The standard deviation is calculated as shown below. Write this out or compare it with equation 4-2 in the textbook to make sure you understand it.

```python
std_dev = np.sqrt(np.sum((x - mean)**2)/(len(x)-1))
print(std_dev)
```

**Part 2: Linear regression.**

Another common thing analytical chemists need to do is generate a calibration curve, such as for a pH meter. 

We will start by simulating a linear response of voltage (in mV) to a known pH input. We will add in some noise using the same function above.

We are using the xlabel() and ylabel() functions to write the axes labels.

```python
slope = 59.16
intercept = 100

x = np.arange(0,10)
y = slope * x + intercept
noise = np.random.normal(0, 10, len(x))
y = y + noise

plt.plot(x, y)
plt.xlabel("pH")
plt.ylabel("Voltage (mV)")
plt.show()
```

Now, imagine we didn't know the slope and the intercept. We need to calculate this from the x and y data. 

To do this, we will use numpy's polyfit command. It can fit any order polynomial, but here we will use an order of 1 to get a linear fit. It returns an array (fit) of the polynomial coefficients in decending order. Thus, the slope is the first value in the array, and the intercept is the second value. These should be close to the values above. If you were to remove the noise (for example by commenting out the line that added it in or by reducing the standard deviation of the noise to 0), you should have values that perfectly match the inputs.

To access individual values from an array, we use []. In computer programming, the first value is actually position 0. So, fit[0] gives you the slope or first element of array. Using fit[1] will extract the second value of the array. 

```python
fit = np.polyfit(x,y, 1)
print(fit)

fit_intercept = fit[1]
fit_slope = fit[0]

fit_ydata = fit_slope * x + fit_intercept

plt.plot(x, y, marker="o")
plt.plot(x, fit_ydata, color="red")
plt.xlabel("pH")
plt.ylabel("Voltage (mV)")
plt.show()
```
